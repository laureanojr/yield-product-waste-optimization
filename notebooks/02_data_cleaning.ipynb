```python
import pandas as pd

df = pd.read_csv("../data/raw/Bakery sales.csv")
print("Starting rows:", len(df))
```

```python
df = df.drop(columns=["Unnamed: 0"])
df.columns
```

```python
df["unit_price"] = (
    df["unit_price"]
    .str.replace("€", "", regex=False)   # remove the euro symbol
    .str.replace(",", ".", regex=False)  # French comma decimal -> dot
    .str.strip()                         # trim spaces
    .astype(float)                       # text -> real number
)
df["unit_price"].describe()
```

```python
df["datetime"] = pd.to_datetime(df["date"] + " " + df["time"])
df[["date", "time", "datetime"]].head()
```

```python
df["ticket_number"] = df["ticket_number"].astype("int64")
df["ticket_number"].head()
```

```python
before = len(df)
df = df[df["Quantity"] > 0]
print(f"Removed {before - len(df)} return rows")
```

```python
before = len(df)
df = df[df["unit_price"] > 0]
print(f"Removed {before - len(df)} zero-price rows")
```

```python
df["revenue"] = df["Quantity"] * df["unit_price"]
df[["article", "Quantity", "unit_price", "revenue"]].head()
```

```python
print("Final rows:", len(df))
print("Unique products:", df["article"].nunique())
print("Date range:", df["datetime"].min(), "->", df["datetime"].max())
df.info()
```

```python
df.to_csv("../data/cleaned/bakery_sales_clean.csv", index=False)
print("Saved:", len(df), "rows to data/cleaned/bakery_sales_clean.
```
